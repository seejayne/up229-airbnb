```python
import pandas as pd
import geopandas as gpd
import contextily as cx
```

# Neighborhood Data

```python
zipfile = "data/Gent-Neighborhoods/Gent-Neighborhoods.shp"
gents_gdf = gpd.read_file(zipfile)
```

```python
gents_gdf.head()
```

```python
gents_gdf.rename(columns={'Gent Score':'Score'}, inplace=True)
gents_gdf.head()
```

```python
# plot the neighborhoods
gents_gdf.plot(color="red", figsize=(9, 9))
```

```python
gents_gdf.info()
```

```python
topgents_gdf=gents_gdf.loc[gents_gdf['Score']>=10].copy() 
topgents_gdf.head()
```

```python
# plot the top neighborhoods
ax = topgents_gdf.plot(color="deeppink",alpha=.5, figsize=(9, 9))
cx.add_basemap(ax, crs=topgents_gdf.crs.to_string(), source=cx.providers.Stamen.TonerLite)
```

# Airbnb Data

```python
# bring in the CSV and take a look at it
df = pd.read_csv('data/airbnb-listings.csv')
df.head(10)
```

```python
df.info()
```

```python
list_gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.longitude, df.latitude))
list_gdf.crs = "epsg:3857"
ax = list_gdf.plot(color="deeppink", alpha=.10, figsize=(9, 9), markersize=4)
cx.add_basemap(ax, crs=topgents_gdf.crs.to_string(), source=cx.providers.Stamen.TonerLite)
```

```python
# only use the listings in the gentrifying areas
from geopandas.tools import sjoin

topgents_gdf.crs = "epsg:3857"

join_list_gent = sjoin(list_gdf, topgents_gdf, how="inner")
join_list_gent.info()
```

```python
print(join_list_gent.crs)
```

```python
# map out the points
ax = join_list_gent.plot(color="deeppink", alpha=.25,figsize=(20, 20), markersize=4)
cx.add_basemap(ax, crs='EPSG:4326', source=cx.providers.Stamen.TonerLite)
```

```python
join_list_gent.to_csv ('data/join_list_gent.csv', index = False, header=True)
```
