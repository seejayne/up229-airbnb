```python
top_N = 10

txt = l.description.str.lower().str.replace(r'\|', ' ').str.cat(sep=' ')
words = nltk.tokenize.word_tokenize(txt)
word_dist = nltk.FreqDist(words)

stopwords = nltk.corpus.stopwords.words('english')
newStopWords = ['<','>','/','br',',','.','b','/b','space','room','!','private','access','kitchen',')','guest','(','&','bedroom','house','apartment','home','bed','2','living','located','-','bathroom',"*",'one','\'s',':','full','la','note','1','large']
stopwords.extend(newStopWords)


words_except_stop_dist = nltk.FreqDist(w for w in words if w not in stopwords) 

rslt = pd.DataFrame(words_except_stop_dist.most_common(top_N),
                    columns=['Word', 'Frequency']).set_index('Word')

matplotlib.style.use('ggplot')

ax = rslt.plot.bar(stacked=True)
```

```python
top_N = 20
newStopWords= ['<','>','/','br',',','.','b','/b','space','room','!','private','access','kitchen',')','guest','(','&','bedroom','house','apartment','home','bed','2','living','located','-','bathroom',"*",'one','\'s',':','full','la','note','1','large','\'','\‘','\’','\'m','—','los angeles', 'would', 'us']
```

```python
def wordbarchart(txt):
    words = nltk.tokenize.word_tokenize(txt)
    word_dist = nltk.FreqDist(words)
    stopwords = nltk.corpus.stopwords.words('english')
    stopwords.extend(newStopWords)
    words_except_stop_dist = nltk.FreqDist(w for w in words if w not in stopwords) 
    rslt = pd.DataFrame(words_except_stop_dist.most_common(top_N),
                        columns=['Word', 'Frequency']).set_index('Word')
    matplotlib.style.use('ggplot')
    ax = rslt.plot.bar(color="deeppink", stacked=True)
```

```python
txt = rv.comments.str.lower().str.replace(r'\|', ' ').str.cat(sep=' ')    
wordbarchart(txt)
```

# Filter review CSV (gentrifying areas)

```python
#pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
#pd.set_option('display.min_rows', None)
```

```python
listDf = pd.read_csv('data/airbnb-listings.csv')
```

```python
rv.head()
```

```python
listDf.head()
```

```python
listDf.set_index('id', inplace=True)
rv.set_index('listing_id', inplace=True)
```

```python
joinedDf = rv.join(listDf, how='left')
```

```python
len(joinedDf)
```

```python
joinedDf.head(3)
```

```python
joinedDf = joinedDf[['id', 'date', 'reviewer_id', 'comments', 'listing_url', 'last_scraped', 'description', 'neighborhood_overview', 'neighbourhood_cleansed', 'host_id', 'latitude', 'longitude']]
```

```python
joinedGdf = gpd.GeoDataFrame(joinedDf, geometry=gpd.points_from_xy(joinedDf.longitude, joinedDf.latitude, crs='EPSG:4326'))
```

```python
#joinedGdf.to_csv('data/reviews_geometry.csv', index = False, header=True)
```

```python
zipfile = "data/Gent-Neighborhoods/Gent-Neighborhoods.shp"
gents_gdf = gpd.read_file(zipfile)
```

```python
gents_gdf.head()
```

```python
gents_gdf.rename(columns={'Gent Score':'Score'}, inplace=True)
gents_gdf.head()
```

```python
# plot the neighborhoods
gents_gdf.plot(color="red", figsize=(9, 9))
```

```python
gents_gdf.info()
```

```python
topgents_gdf=gents_gdf.loc[gents_gdf['Score']>=10].copy() 
topgents_gdf.head()
```

```python
# plot the top neighborhoods
ax = topgents_gdf.plot(color="deeppink",alpha=.5, figsize=(9, 9))
cx.add_basemap(ax, crs=topgents_gdf.crs.to_string(), source=cx.providers.Stamen.TonerLite)
```

```python
joinedGdf.crs = "epsg:3857"
ax = joinedGdf.plot(color="deeppink", alpha=.10, figsize=(9, 9), markersize=4)
cx.add_basemap(ax, crs=topgents_gdf.crs.to_string(), source=cx.providers.Stamen.TonerLite)
```

```python
# only use the listings in the gentrifying areas
from geopandas.tools import sjoin

topgents_gdf.crs = "epsg:3857"

join_rv_gent = sjoin(joinedGdf, topgents_gdf, how="inner")
join_rv_gent.info()
```

```python
ax = join_rv_gent.plot(color="deeppink", alpha=.25,figsize=(20, 20), markersize=4)
cx.add_basemap(ax, crs='EPSG:4326', source=cx.providers.Stamen.TonerLite)
```

```python
join_rv_gent.to_csv('data/join_rv_gent2.csv', index = False, header=True)
```
