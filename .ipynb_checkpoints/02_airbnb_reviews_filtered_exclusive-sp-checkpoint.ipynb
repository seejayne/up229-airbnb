```python
import pandas as pd
import geopandas as gpd
import contextily as cx
```

```python
#pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
#pd.set_option('display.min_rows', None)
```

```python
listDf = pd.read_csv('data/airbnb-listings.csv')
```

```python
rv = pd.read_csv('data/reviews.csv')
```

```python
rv.head()
```

```python
listDf.head()
```

```python
listDf.set_index('id', inplace=True)
rv.set_index('listing_id', inplace=True)
```

```python
joinedDf = rv.join(listDf, how='left')
```

```python
len(joinedDf)
```

```python
joinedDf.head(3)
```

```python
joinedDf = joinedDf[['id', 'date', 'reviewer_id', 'comments', 'listing_url', 'last_scraped', 'description', 'neighborhood_overview', 'neighbourhood_cleansed', 'host_id', 'latitude', 'longitude']]
```

```python
joinedGdf = gpd.GeoDataFrame(joinedDf, geometry=gpd.points_from_xy(joinedDf.longitude, joinedDf.latitude, crs='EPSG:4326'))
```

```python
#joinedGdf.to_csv('data/reviews_geometry.csv', index = False, header=True)
```

```python
zipfile = "data/Exclusive Neighborhoods/Exclusive-Neighborhoods.shp"
exclusive_gdf = gpd.read_file(zipfile)
```

```python
exclusive_gdf.head()
```

```python
# plot the neighborhoods
exclusive_gdf.plot(color="blue", figsize=(9, 9))
```

```python
exclusive_gdf.info()
```

```python
# plot the top neighborhoods
ax = exclusive_gdf.plot(color="blue",alpha=.5, figsize=(9, 9))
cx.add_basemap(ax, crs=exclusive_gdf.crs.to_string(), source=cx.providers.Stamen.TonerLite)
```

```python
joinedGdf.crs = "epsg:3857"
ax = joinedGdf.plot(color="blue", alpha=.10, figsize=(9, 9), markersize=4)
cx.add_basemap(ax, crs=exclusive_gdf.crs.to_string(), source=cx.providers.Stamen.TonerLite)
```

```python
#list_gdf = gpd.GeoDataFrame(
#    listDf, geometry=gpd.points_from_xy(listDf.longitude, listDf.latitude))
#list_gdf.crs = "epsg:3857"
#ax = list_gdf.plot(color="blue", alpha=.10, figsize=(9, 9), markersize=4)
#cx.add_basemap(ax, crs=exclusive_gdf.crs.to_string(), source=cx.providers.Stamen.TonerLite)
```

```python
# only use the listings in the exclusive areas
from geopandas.tools import sjoin

exclusive_gdf.crs = "epsg:3857"

join_rv_exclusive = sjoin(joinedGdf, exclusive_gdf, how="inner")
join_rv_exclusive.info()
```

```python
ax = join_rv_exclusive.plot(color="blue", alpha=.25,figsize=(20, 20), markersize=4)
cx.add_basemap(ax, crs='EPSG:4326', source=cx.providers.Stamen.TonerLite)
```

```python
print(join_rv_exclusive.crs)
```

```python
join_rv_exclusive.to_csv ('data/join_rv_exclusive.csv', index = False, header=True)
```

# Airbnb Data

```python
# bring in the CSV and take a look at it
df = pd.read_csv('data/airbnb-listings.csv')
df.head(10)
```

```python
df.info()
```

```python
list_gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.longitude, df.latitude))
list_gdf.crs = "epsg:3857"
ax = list_gdf.plot(color="blue", alpha=.10, figsize=(9, 9), markersize=4)
cx.add_basemap(ax, crs=exclusive_gdf.crs.to_string(), source=cx.providers.Stamen.TonerLite)
```

```python
# only use the listings in the exclusive areas
from geopandas.tools import sjoin

exclusive_gdf.crs = "epsg:3857"

join_list_exclusive = sjoin(list_gdf, exclusive_gdf, how="inner")
join_list_exclusive.info()
```

```python
print(join_list_exclusive.crs)
```

```python
# map out the points
ax = join_list_exclusive.plot(color="blue", alpha=.25,figsize=(20, 20), markersize=4)
cx.add_basemap(ax, crs='EPSG:4326', source=cx.providers.Stamen.TonerLite)
```

```python
join_list_exclusive.to_csv ('data/join_list_exclusive.csv', index = False, header=True)
```
